```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
ver = 'Masks_S_v5'
```

```python
xs = []
# ys = []
names = []
for f in os.listdir(f'/Users/reza/Career/DMLab/DKIST/frontiers_repo/data/{ver}/Train'):
    a = np.load(f'/Users/reza/Career/DMLab/DKIST/frontiers_repo/data/{ver}/Train/{f}')
    names.append(f)
    x = a['smap']
    y = a['cmask_map']
    xs.append(x)
    plt.figure(figsize=(10, 5))
    plt.title(f)
    plt.subplot(1, 2, 1)
    plt.imshow(x, cmap='gray')
    plt.subplot(1, 2, 2)
    plt.imshow(y, cmap=plt.get_cmap('PiYG', 5))
    plt.show()

for f in os.listdir(f'/Users/reza/Career/DMLab/DKIST/frontiers_repo/data/{ver}/Validate'):
    a = np.load(f'/Users/reza/Career/DMLab/DKIST/frontiers_repo/data/{ver}/Validate/{f}')
    names.append(f)
    x = a['smap']
    y = a['cmask_map']
    xs.append(x)
    plt.figure(figsize=(10, 5))
    plt.title(f)
    plt.subplot(1, 2, 1)
    plt.imshow(x, cmap='gray')
    plt.subplot(1, 2, 2)
    plt.imshow(y, cmap=plt.get_cmap('PiYG', 5))
    plt.show()
    
```

```python
for i, x in enumerate(xs):
    for ii, xx in enumerate(xs):
        if i == ii:
            continue
        duplicate = (x==xx).all()
        if duplicate:
            print(i, ii)
            plt.figure(figsize=(10, 5))
            plt.subplot(1, 2, 1)
            plt.title(names[i])
            plt.imshow(x, cmap='gray')
            plt.subplot(1, 2, 2)
            plt.title(names[ii])
            plt.imshow(xx, cmap='gray')
```
